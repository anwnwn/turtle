```python
# 10 days periodic Analysis
data['Momentum_10D'] = data['Close'].diff(periods=10) 
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Momentum_10D'], label='10-Day Momentum', color='mediumorchid')
plt.title('10-Day Momentum of IPC MEXICO Closing Prices')
plt.xlabel('Date')
plt.ylabel('Momentum')
plt.xticks(data['Date'][::len(data)//10], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Mean Reversion
```

```python
data['50D_MA'] = data['Close'].rolling(window=50).mean()
buy_threshold = 0.95
sell_threshold = 1.00
data['Signal'] = 0 

for i in range(1, len(data)):
    if data['Close'].iloc[i] < buy_threshold * data['50D_MA'].iloc[i] and data['Signal'].iloc[i-1] != 1:
        data['Signal'].iloc[i] = 1  
    elif data['Close'].iloc[i] > sell_threshold * data['50D_MA'].iloc[i] and data['Signal'].iloc[i-1] == 1:
        data['Signal'].iloc[i] = -1 

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Closing Price', alpha=0.5, color='blue')
plt.plot(data['Date'], data['50D_MA'], label='50-Day Moving Average', color='gold', alpha=0.75)

plt.scatter(data.loc[data['Signal'] == 1, 'Date'], data.loc[data['Signal'] == 1, 'Close'], label='Buy Signal', marker='^', color='green', alpha=1)
plt.scatter(data.loc[data['Signal'] == -1, 'Date'], data.loc[data['Signal'] == -1, 'Close'], label='Sell Signal', marker='v', color='red', alpha=1)

plt.title('Mean Reversion Strategy: Buy and Sell Signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(data['Date'][::len(data)//10], rotation=45)  
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate the short-term (5-day) and long-term (20-day) simple moving averages (SMA) of the closing prices
data['SMA5'] = data['Close'].rolling(window=5).mean()
data['SMA20'] = data['Close'].rolling(window=20).mean()

data['MR Ratio'] = data['SMA5'] / data['SMA20']

plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['MR Ratio'], label='MR Ratio (SMA5/SMA20)', color='blue')

plt.axhline(y=1, color='r', linestyle='--', label='Mean Reversion Level')
plt.title('Mean Reversion (MR) Ratio Over Time')
plt.xlabel('Date')
plt.ylabel('MR Ratio')
plt.legend()
plt.grid(True)
plt.show()

mr_rtn = data['MR Ratio'].describe()
mr_rtn
```
